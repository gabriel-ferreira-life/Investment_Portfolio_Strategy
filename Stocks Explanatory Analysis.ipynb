## Apple Bollinger Bands

```python
# Subsetting the data to have a better visualization of the bollinger bands
AAPL_bollinger = AAPL[(AAPL.index >= '2020-01-01')]


                         
# Computing the moving average and standard deviation
AAPL_bollinger['ma'] = AAPL_bollinger.loc[:,'Close'].rolling(window = 20).mean()
AAPL_bollinger['sd'] = AAPL_bollinger.loc[:,'Close'].rolling(window = 20).std(doof = 1)

# Computing the Bollinger Bands
AAPL_bollinger['up'] = AAPL_bollinger.loc[:,'ma'] + 2 * AAPL_bollinger['sd']
AAPL_bollinger['low'] = AAPL_bollinger.loc[:,'ma'] - 2 * AAPL_bollinger['sd']

# Create a function to get the buy and sell signals
def get_signal(data):
    buy_signal = [] #buy list
    sell_signal = [] #sell list
    for i in range(len(AAPL_bollinger['Close'])):
        if AAPL_bollinger['Close'][i] > AAPL_bollinger['up'][i]: #Then you should sell 
      #print('SELL')
            buy_signal.append(np.nan)
            sell_signal.append(AAPL_bollinger['Close'][i])
        elif AAPL_bollinger['Close'][i] < AAPL_bollinger['low'][i]: #Then you should buy
      #print('BUY')
            sell_signal.append(np.nan)
            buy_signal.append(AAPL_bollinger['Close'][i])
        else:
            buy_signal.append(np.nan)
            sell_signal.append(np.nan)
    return (buy_signal, sell_signal)


#Create new columns for the buy and sell signals
AAPL_bollinger['Buy'] =  get_signal(AAPL_bollinger)[0]
AAPL_bollinger['Sell'] =  get_signal(AAPL_bollinger)[1]

# Plotting the Bollinger Bands
plt.figure(figsize=(15, 10))
x_axis = AAPL_bollinger.index
plt.plot(AAPL_bollinger['Close'], color = 'black', label = 'Close Price', lw = 1)
plt.plot(AAPL_bollinger['ma'], color = 'green', label = '20-day MA', lw = 1)
plt.plot(AAPL_bollinger['up'], color = 'blue', label = 'Upper Band', lw = 1)
plt.plot(AAPL_bollinger['low'], color = 'blue', label = 'Lower Band', lw = 1)
plt.plot(x_axis, AAPL_bollinger['Buy'], color = 'green', lw = 1, label = 'Buy',marker = '^', markersize = 10)
plt.plot(x_axis, AAPL_bollinger['Sell'], color = 'red', lw = 1, label = 'Sell',marker = 'v', markersize = 10)
plt.title('Apple - Bollinger Bands')
plt.ylabel('Apple Close Price ($)')
plt.xlabel('Date')
plt.legend(loc = 'upper left')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```

## Momentum: Relative Strength Index (RSI) - Apple

```python
# Calculating difference in price
AAPL['Delta'] = AAPL['Close'].diff()

# Computing dummy variables to indicate whether price went up or down
AAPL['up'] = np.where(AAPL['Delta'] > 0, 1, 0)
AAPL['down'] = np.where(AAPL['Delta'] <= 0, 1, 0)

# Dropping missings values
AAPL = AAPL.dropna(axis = 0)

# Computing prices for up and down
AAPL['up_val'] = abs(AAPL['Delta'] * AAPL['up'])
AAPL['down_val'] = abs(AAPL['Delta'] * AAPL['down'])

# Computing the up and down averages
AAPL['up_avg'] = AAPL['up_val'].rolling(window = 14).mean()
AAPL['down_avg'] = AAPL['down_val'].rolling(window = 14).mean()

# Computing the up and down Wilder Moving Averages
AAPL['up_WMA'] = AAPL['up_avg']
AAPL['down_WMA'] = AAPL['down_avg']

for i in range(14, AAPL.shape[0]):
        AAPL['up_WMA'][i] = AAPL['up_WMA'][i-1] + (AAPL['up_avg'][i] - AAPL['up_WMA'][i-1]) / 14
        AAPL['down_WMA'][i] = AAPL['down_WMA'][i-1] + (AAPL['down_avg'][i] - AAPL['down_WMA'][i-1]) / 14

# Calculating the RS
AAPL['RS'] = AAPL['up_WMA'] / AAPL['down_WMA']

# Calculating the RSI
AAPL['RSI'] = 100 - (100/ (1 + AAPL['RS'])) 

# Visualizing the RSI
plt.figure(figsize=(15, 10))
plt.plot('RSI', data = AAPL, color = 'black')
plt.axhline(y = 30, color = 'black', linestyle = '--')
plt.axhline(y = 70, color = 'black', linestyle = '--')
plt.title('Momentum: Relative Strength Index (RSI) - Apple')
plt.ylabel('RSI (14-day Moving Average)')
plt.xlabel('Date')
plt.xticks(rotation = 45)
plt.ylim(25, 85)
plt.grid()
```

```python
# Calculating difference in price
FB['Delta'] = FB['Close'].diff()

# Computing dummy variables to indicate whether price went up or down
FB['up'] = np.where(FB['Delta'] > 0, 1, 0)
FB['down'] = np.where(FB['Delta'] <= 0, 1, 0)

# Dropping missings values
FB = FB.dropna(axis = 0)

# Computing prices for up and down
FB['up_val'] = abs(FB['Delta'] * FB['up'])
FB['down_val'] = abs(FB['Delta'] * FB['down'])

# Computing the up and down averages
FB['up_avg'] = FB['up_val'].rolling(window = 14).mean()
FB['down_avg'] = FB['down_val'].rolling(window = 14).mean()

# Computing the up and down Wilder Moving Averages
FB['up_WMA'] = FB['up_avg']
FB['down_WMA'] = FB['down_avg']

for i in range(14, FB.shape[0]):
        FB['up_WMA'][i] = FB['up_WMA'][i-1] + (FB['up_avg'][i] - FB['up_WMA'][i-1]) / 14
        FB['down_WMA'][i] = FB['down_WMA'][i-1] + (FB['down_avg'][i] - FB['down_WMA'][i-1]) / 14

# Calculating the RS
FB['RS'] = FB['up_WMA'] / FB['down_WMA']

# Calculating the RSI
FB['RSI'] = 100 - (100/ (1 + FB['RS']))

# Visualizing the RSI
plt.figure(figsize=(15, 10))
plt.plot('RSI', data = FB, color = 'black')
plt.axhline(y = 30, color = 'black', linestyle = '--')
plt.axhline(y = 70, color = 'black', linestyle = '--')
plt.title('Momentum: Relative Strength Index (RSI) - Facebook')
plt.ylabel('RSI (14-day Moving Average)')
plt.xlabel('Date')
plt.xticks(rotation = 45)
plt.ylim(25, 85)
plt.grid()
```

## Microsoft Bollinger Bands

```python
# Subsetting the data to have a better visualization of the bollinger bands
MSFT_bollinger = MSFT[(MSFT.index >= '2020-01-01')]

# Computing the moving average and standard deviation
MSFT_bollinger['ma'] = MSFT_bollinger['Close'].rolling(window = 20).mean()
MSFT_bollinger['sd'] = MSFT_bollinger['Close'].rolling(window = 20).std(doof = 1)

# Computing the Bollinger Bands
MSFT_bollinger['up'] = MSFT_bollinger['ma'] + 2 * MSFT_bollinger['sd']
MSFT_bollinger['low'] = MSFT_bollinger['ma'] - 2 * MSFT_bollinger['sd']

# Create a function to get the buy and sell signals
def get_signal(data):
    buy_signal = [] #buy list
    sell_signal = [] #sell list
    for i in range(len(MSFT_bollinger['Close'])):
        if MSFT_bollinger['Close'][i] > MSFT_bollinger['up'][i]: #Then you should sell 
      #print('SELL')
            buy_signal.append(np.nan)
            sell_signal.append(MSFT_bollinger['Close'][i])
        elif MSFT_bollinger['Close'][i] < MSFT_bollinger['low'][i]: #Then you should buy
      #print('BUY')
            sell_signal.append(np.nan)
            buy_signal.append(MSFT_bollinger['Close'][i])
        else:
            buy_signal.append(np.nan)
            sell_signal.append(np.nan)
    return (buy_signal, sell_signal)

#Create new columns for the buy and sell signals
MSFT_bollinger['Buy'] =  get_signal(MSFT_bollinger)[0]
MSFT_bollinger['Sell'] =  get_signal(MSFT_bollinger)[1]

# Plotting the Bollinger Bands
plt.figure(figsize=(15, 10))
x_axis = MSFT_bollinger.index
plt.plot(MSFT_bollinger['Close'], color = 'black', label = 'Close Price', lw = 1)
plt.plot(MSFT_bollinger['ma'], color = 'blue', label = '20-day MA', lw = 1)
plt.plot(MSFT_bollinger['up'], color = 'green', label = 'Upper Band', lw = 1)
plt.plot(MSFT_bollinger['low'], color = 'green', label = 'Lower Band', lw = 1)
plt.plot(x_axis, MSFT_bollinger['Sell'], color = 'red', lw = 1, label = 'Sell',marker = 'v', markersize = 10)
plt.plot(x_axis, MSFT_bollinger['Buy'], color = 'green', lw = 1, label = 'Buy',marker = '^', markersize = 10)
plt.title('Microsoft - Bollinger Bands')
plt.ylabel('Microsoft Close Price ($)')
plt.xlabel('Date')
plt.legend(loc = 'upper left')
plt.xticks(rotation = 45)
plt.grid()
plt.show()
```
